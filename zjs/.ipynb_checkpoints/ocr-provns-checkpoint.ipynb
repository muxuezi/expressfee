```python
import collections
```

```python
temp = []
def textset(fid):
    idx = 0
    f = open('zjs%d' % fid)
    lab = collections.Counter()
    for x in f.readlines():
        head = x.split('-')[0]
        if lab[head] < 1 : 
            temp.append(x)
            print idx, x[:-1]
            idx += 1
        lab[head] +=1
map(textset,range(5))
```
