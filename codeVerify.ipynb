```python
class Codecutting(object):

    def __init__(self, fname, width, height, marginleft, margintop, interval):
        self.fname = fname
        self.ffont = fname + 'font'
        self.width = width
        self.height = height
        self.marginleft = marginleft
        self.margintop = margintop
        self.interval = interval

    def cutascii(self, f):
        f = "./" + self.fname + '/' + f
        im = Image.open(f)
#         im = im.filter(ImageFilter.MedianFilter())
#         enhancer = ImageEnhance.Contrast(im)
#         im = enhancer.enhance(2)
        im = im.convert('L')
        im.show()
        threshold = 153
        table = []
        
        for i in range(256):
            if i > threshold:
                table.append(0)
            else:
                table.append(1)
            
        im = im.point(table,'1') 
        im.show()
        
        for i in range(4):
            x = marginleft + i * (width + interval)
            y = margintop
            im.crop(
                (x, y, x + width, y + height)).save("./" + self.ffont + "/%04d.bmp" %
                                                    random.randrange(10000))

    def readfile(self):
        return filter(lambda f: f.endswith(".aspx"), os.listdir("./" + self.fname))

    def cutall(self):
        map(lambda f: self.cutascii(f), self.readfile()[:2])

fname = 'yt'
width = 8
height = 12
marginleft = 6
margintop = 6
interval = 1

ytcut = Codecutting(fname, width, height, marginleft, margintop,interval)
# ytcut.readfile()
ytcut.cutall()
```

```python
# load font modules  (char, image)
fontMods = []
for i in range(10):
    fontMods.append((str(i), Image.open("./good/%02d.bmp" % i)))
for i in range(26):
    c = chr(ord('A') + i)
    fontMods.append((c, Image.open("./good/%s.bmp" % c)))

def recognize(f):
    im = Image.open(f)
    im2 = im.convert('1')
    # check 5 fonts
    result = "./result/"
    for i in range(5):
        x = 10 + i*18
        y = 6
        target = im.crop((x, y, x+8, y+10))
        points = []
        for mod in fontMods:
            diffs = 0
            for yi in range(10):
                for xi in range(8):
                    if mod[1].getpixel((xi, yi)) != target.getpixel((xi, yi)):
                        diffs += 1
            points.append((diffs, mod[0]))
        points.sort()
        result += points[0][1]
    result += ".png"
    print "save to", result
    im.save(result);

for imgfile in os.listdir("."):
    if imgfile.endswith(".png"):
        recognize(imgfile)
```
