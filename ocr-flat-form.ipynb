```python
# coding: utf-8
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.common.exceptions import NoSuchElementException
import time
import os
from PIL import Image
```

```python
class FlatList(object):
    def __init__(self,url,outname,provnid,cityid,areaid):
        self.url = url
        self.outname = outname
        self.provnid = provnid
        self.cityid = cityid
        self.areaid = areaid
        
    def flat_selection(self):
        driver = webdriver.Chrome('./chromedriver')
        driver.implicitly_wait(30)
        driver.get(self.url)
        with open(self.outname, 'w') as f:
            provns = Select(driver.find_element_by_id(self.provnid))
            provn_len = len(provns.options)
#             provn_len = 21
            prof,prot = 1, provn_len # zjs
            prof,prot = 0, provn_len # zhongtong
#             prof,prot = 3,provn_len-2   # yt
#             prof,prot = provn_len-2,provn_len  # yt
            for idx in range(prof,prot):
                provns.select_by_index(idx)
                provns_text = provns.first_selected_option.text
                ################### shunfeng, zhongtong 
                allarea = []
                cities = Select(driver.find_element_by_id(self.cityid))
#                     time.sleep(1)
                cities = driver.find_element_by_id(self.cityid)
                allcity = map(lambda x:x.text,cities.find_elements_by_tag_name('option'))
                item = lambda cities_text: (provns_text.encode('utf-8'),cities_text.encode('utf-8'), 'None')
                map(lambda x: f.write('-'.join(item(x)) + '\n'), allcity)
                allarea = []
                    ################### shunfeng, zhongtong 
                    
#                 cities = Select(driver.find_element_by_id(self.cityid))
#                 cities_len = len(cities.options) 
#                 allarea = []
#                 for cidx in range(1, cities_len):
#                     cities = Select(driver.find_element_by_id(self.cityid))
#                     cities.select_by_index(cidx)
#                     cities_text = cities.first_selected_option.text
# #                     time.sleep(1)
#                     areas = driver.find_element_by_id(self.areaid)
#                     allarea = map(lambda x:x.text,areas.find_elements_by_tag_name('option'))
#                     item = lambda areatemp: (provns_text.encode('utf-8'),cities_text.encode('utf-8'), areatemp)
#                     if len(allarea) > 1:map(lambda x: f.write('-'.join(item(x)) + '\n'), allarea[1:])
#                     else:f.write('-'.join(item('None')) + '\n')
#                     allarea = []

# url = "http://www.zjs.com.cn/WS_Business/WS_Business_price_internal.aspx?id=6"
# outname = 'zjsdataonly0.txt'
# provnid = 'ctl00_Content1_Price_Province'
# cityid = 'City'
# areaid = 'Area'
# flatdst = FlatList(url,outname,provnid,cityid,areaid)
# flatdst.flat_selection()
# url = "http://www.yto.net.cn/cn/service/standardPrice.html"
# outname = 'yuantongdata.txt'
# provnid = 'selPro'
# cityid = 'selCity'
# areaid = 'selDis'
# flatdst = FlatList(url,outname,provnid,cityid,areaid)
# flatdst.flat_selection()
url = "http://www.zto.cn/Price.aspx"
outname = 'zhongtongdata.txt'
provnid = 's_province_text'
cityid = 's_city_text'
areaid = 's_city_text'
flatdst = FlatList(url,outname,provnid,cityid,areaid)
flatdst.flat_selection()
```
